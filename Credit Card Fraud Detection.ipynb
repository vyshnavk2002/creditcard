```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('/content/creditcard.csv')
data.head()
```

data into legit & fraud transactions

```python
legit = data[data.Class==0]
fraud = data[data.Class==1]
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
legit_sample=legit.sample(n=492)
```

Concatinating 2 datasets

```python
new_data_set=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data_set.head()
```

```python
new_data_set.tail()
```

```python
new_data_set['Class'].value_counts()
```

```python
new_data_set.groupby('Class').mean()
```

Splitting data into features & targets
features=v1,v2.....
targets=0,1

```python
x=new_data_set.drop(columns='Class',axis=1)
y=new_data_set['Class']
```

```python
print(x)
```

```python
print(y)
```

Split data into training & testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.2,stratify=y)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

training logistic regression model with training data


```python
model.fit(x_train,y_train)
```

Model evaluation

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print('Accuracy on Test data:',test_data_accuracy)
```
